```python
import torch
from torch import nn
from torch.nn import functional as F
import numpy as np
import matplotlib.pyplot as plt
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
from tqdm.notebook import tqdm
from funcs import DDPM, training_loop_ddpm, generate_image, show_images, return_dataloader
```

UNetModel

```python
from guided.guided_diffusion.unet import UNetModel
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
network = UNetModel(
    image_size=32,
    in_channels=1,
    model_channels=32,
    out_channels=1,
    num_res_blocks=2,
    attention_resolutions=(2, 2),
    channel_mult=(1, 2, 4, 8),
    num_classes=10
)
```

```python
dataloader = return_dataloader()
```

```python
learning_rate = 1e-3
num_epochs = 50
num_timesteps = 1000

network = network.to(device)
model = DDPM(network, num_timesteps, beta_start=0.0001, beta_end=0.02, device=device)
optimizer = torch.optim.Adam(network.parameters(), lr=learning_rate)
training_loop_ddpm(model, dataloader, optimizer, num_epochs, num_timesteps, device=device)
```

```python
torch.save(model.state_dict(), "ddpm_mnist.pth")
```
